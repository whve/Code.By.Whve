```python
# 导入snp csv,生成ndarry数组
import numpy as np
import pandas as pd
hrj = pd.read_csv(r'E:\whve.AD.blood\yavbyj\ya.tsv',delimiter = '\t')

pd.set_option('display.max_columns', None)

hrj2=hrj.loc[:, ['SNPS', 'OR or BETA','95% CI (TEXT)','P-VALUE','DISEASE/TRAIT','MAPPED_TRAIT','PUBMEDID','MAPPED_GENE','STUDY ACCESSION']]
hrj2
```

```python
hrj3=hrj2[hrj2['P-VALUE']<5*10**-8]
hrj3
```

```python
jib = pd.read_csv(r'F:\AD\[01]AD\IGAP_stage_1.txt',delimiter = '\t')
# AD data set
jib2=jib.loc[:, ['MarkerName', 'Effect_allele','Non_Effect_allele','Beta','SE','Pvalue']]
# 重命名
jib21=jib2.rename(columns={'MarkerName':'SNPS'})
hvzs=pd.merge(hrj3,jib21,how='inner',on='SNPS')
hvzs
```

```python
hvzs2=hvzs.loc[:, ['STUDY ACCESSION','SNPS', 'OR or BETA','95% CI (TEXT)','Effect_allele','Non_Effect_allele','P-VALUE','DISEASE/TRAIT','MAPPED_TRAIT','PUBMEDID','MAPPED_GENE','STUDY ACCESSION','Beta','SE','Pvalue']]

hvzs3=hvzs2.rename(columns={'STUDY ACCESSION':'Phenotype','SNPS':'SNP','OR or BETA':'beta','95% CI (TEXT)':'se','Effect_allele':'effect_allele','Non_Effect_allele':'other_allele'})
hvzs3
```
