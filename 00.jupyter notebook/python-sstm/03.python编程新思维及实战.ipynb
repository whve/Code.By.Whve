```python
def histogram(s):
    ''' 统计字符出现次数'''
    d=dict()
    for c in s:
        if c not in ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']:
            pass
        elif c not in d:
            d[c]=1
        else:
            d[c]+=1
    return d

fin=open("info.txt","r")
'''打开1.txt,求其第一段的字符次数与频率'''
# b="a tenth grade student at the school is one of the students who signed the petition. she said that getting to and from classes in five minutes might be difficult for students who are commuting across campus.a spokesperson for muskogee high school was not immediately available for comment when contacted by fox news on saturday."
b = fin.readline() # readline and readlines
d = (histogram(b))
ls = list(d.items())
ls.sort()
for k, c in ls:
    print("{}:{}".format(k, c))
fin.close()
```

```python
f = open("data.csv", "r", encoding = "utf-8")
ls = []
lt = []
for line in f:
    line = line.strip(" \n")
    if line != "":  #去空行
        t = []
        for i in line.split(","): 
            t.append(i.strip(" "))  #去掉每个元素前后空格
        ls.append(t)
for i in range(len(ls[0])):   #生成二维空列表
    lt.append([])
for i in range(len(ls)-1, -1, -1):   #转置计算
    for j in range(len(ls[i])):
        lt[j].append(ls[i][j])
for item in lt:          
    print(",".join(item))
f.close()
```

```python
import jieba
f = open("白鹿原.txt")
ls = jieba.lcut(f.read())
A = set(ls)
maxw = ""
for w in A:
    if len(w) > len(maxw):
        maxw = w
    if len(w) == len(maxw) and w > maxw:
        maxw = w
print(maxw)
f.close()
```
